```python
avar_wiki_page_names_1 = []
```

```python
def get_pages_from_listpage(url, avar_wiki_page_names_1):
    soup = bs(requests.get(url).text, 'html.parser')
    next_page = soup.find('div', {"class": "mw-allpages-nav"}, 1).find_all('a')[-1]
    for li in soup.find_all('li'):
        if not li.attrs and li.find('a'): avar_wiki_page_names_1.append(li.getText())
    # avar_wiki_page_names += [a.getText() for a in soup.find_all('a', {"class": "mw-redirect"})]
    if next_page.text.startswith('Следующая страница'):
        print('looking ' + 'https://av.wikipedia.org' + next_page.get('href'))
        time.sleep(0.2)
        get_pages_from_listpage('https://av.wikipedia.org' + next_page.get('href'), avar_wiki_page_names_1)
    else:
        print('END with ' + next_page.text)
        print(avar_wiki_page_names_1)
        copy_2cb(avar_wiki_page_names_1)
        return avar_wiki_page_names_1
```

### Работа со списком статей

```python
import wikipedia
from tqdm import tqdm
```

```python
wikipedia.set_lang('av')
```

```python
list_articles = avar_wiki_page_names + avar_wiki_page_names_1
```

```python
res = []
```

```python
wikipedia.WikipediaPage('Дербенд мухъ').content
```

```python
list_1 = list_articles[:1000]
# list_2 = list_articles[1000:2000]
# list_2 = list_articles[2000:3000]
# list_3 = list_articles[3000:4000]
# list_4 = list_articles[4000:5000]
# list_5 = list_articles[5000:]
```

```python
res = []
for i in tqdm(range(len(list_1))):
    try:
        res += re.sub(r'\n+==.*?==', '', wikipedia.WikipediaPage(list_1[i]).content, flags=re.M).split('\n')
    except:
        print('problem: ' + list_1[i])
```

```python
with open('wikipedia_5.json', 'w', encoding='utf-8') as outfile:
    json.dump({'texts': list(filter(None, res))}, outfile, ensure_ascii=False)
```
